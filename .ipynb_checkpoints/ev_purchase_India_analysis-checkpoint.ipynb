#### Loading the laptop_prices dataset using Pandas

```python
file_path = 'ev_dataset.csv'
try:
    # Attempt to read the CSV file
    df = pd.read_csv(file_path)
    print("CSV file loaded successfully!")
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except pd.errors.ParserError:
    print(f"Error: The file '{file_path}' could not be parsed (malformed CSV).")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

#### Grouping Categorical Columns and Calculating Mean and Sorting

##### Grouping By State

```python
# Group by 'State' and calculate the mean of 'EV_Sales_Quantity'
grouped_state = df.groupby(['State', 'Year'])['EV_Sales_Quantity'].mean().sort_values(ascending = False)

# Display the result
print(grouped_state)
```

##### Grouping By Month

```python
# Group by 'Month_Name' and calculate the mean of 'EV_Sales_Quantity'
grouped_month = df.groupby(['Month_Name', 'Year'])['EV_Sales_Quantity'].mean().sort_values(ascending = False)

# Display the result
print(grouped_month)
```

##### Grouping By Vehicle Class

```python
# Group by 'Vehicle_Class' and calculate the mean of 'EV_Sales_Quantity'
grouped_vehicle_class = df.groupby(['Vehicle_Class', 'Year'])['EV_Sales_Quantity'].mean().sort_values(ascending = False)

# Display the result
print(grouped_vehicle_class)
```

##### Grouping By Vehicle Category

```python
# Group by 'Vehicle_Category' and calculate the mean of 'EV_Sales_Quantity'
grouped_vehicle_category = df.groupby(['Vehicle_Category', 'Year'])['EV_Sales_Quantity'].mean().sort_values(ascending = False)

# Display the result
print(grouped_vehicle_category)
```

##### Grouping By Vehicle Type

```python
# Group by 'Vehicle_Type' and calculate the mean of 'EV_Sales_Quantity'
grouped_vehicle_type = df.groupby(['Year', 'Vehicle_Type'])['EV_Sales_Quantity'].mean().sort_values(ascending = False)

# Display the result
print(grouped_vehicle_type)
```

#### Patterns or Interesting Findings

- **Total Sales**: 96,845 units of electric vehicles were sold.
- **Top-Performing State**: Uttar Pradesh recorded the highest average sales with a mean of 160 units in 2024.
- **Top Month**: In May 2023, highest sales were recorded with a mean of 184 units being sold.
- **Top Vehicle Class**: M-CYCLE/SCOOTER topped 2023 and 2024 sales with mean of 2021 and 2314 units respectively. 
- **Top Vehicle Category**: 2-Wheelers recorded highest sales in 2023 and 2024.
- **Top Vehicle Type**: 3W_Shared_LowSpeed was the best-selling vehicle type, with 1775 units sold in 2023.

# Task 3

Create at least four different types of visualizations:
- Line chart showing trends over time (for example, a time-series of sales data).
- Bar chart showing the comparison of a numerical value across categories (e.g., average petal length per species).
- Histogram of a numerical column to understand its distribution.
- Scatter plot to visualize the relationship between two numerical columns (e.g., sepal length vs. petal length).<br>
Customize your plots with titles, labels for axes, and legends where necessary.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Line Chart Showing EV_Sales Trends

```python
# Group data by 'Year' and calculate total EV sales
annual_sales = df.groupby('Year')['EV_Sales_Quantity'].sum().reset_index()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(annual_sales['Year'], annual_sales['EV_Sales_Quantity'], color='blue', marker='o', linestyle='-', label='Annual Sales')

# Add titles and labels
plt.title('Annual EV Sales Trend', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total EV Sales Quantity', fontsize=12)
plt.legend()

# Add grid for better visualization
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

## Bar Chart For Categorical Variables

### Bar Chart For Comparing EV Sales By Year

```python
# Group data by 'Year' and calculate total EV sales
yearly_sales = df.groupby('Year')['EV_Sales_Quantity'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='EV_Sales_Quantity', data=yearly_sales, palette='viridis')

# Add titles and labels
plt.title('Total EV Sales by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total EV Sales Quantity', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

## Histogram for Electric Vehicles Sales Quantity in India

```python
# Filter data between 2014 and 2024
df_filtered = df[(df['Year'] >= 2014) & (df['Year'] <= 2024)]

# Plotting the histogram for EV_Sales_Quantity
plt.figure(figsize=(10, 6))
sns.histplot(df_filtered['EV_Sales_Quantity'], bins=15, kde=True, color='blue')

# Add titles and labels
plt.title('Distribution of EV Sales Quantity (2014-2024)', fontsize=16)
plt.xlabel('EV Sales Quantity', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

## Scatter Plot for Year and EV_Sales_Quantity

```python
# Scatter plot between Year and EV_Sales_Quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='EV_Sales_Quantity', data=df, color='blue')

# Add titles and labels
plt.title('EV Sales Quantity Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('EV Sales Quantity', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

### Findings

1. 2023 recorded the highest sales compared to other years.
2. Lowest sales were recorded in 2014.

## Insights

1. There is an increasing adoption of electric vehicles with 2-Wheelers taking precedence.
2. Incentives and support systems in India need to be improved to support the growing need of electric vehicles.
